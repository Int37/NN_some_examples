```python
#Импорт необходимых  библеотек
import numpy as np 
import matplotlib.pyplot as plt
from tensorflow.keras import layers
from tensorflow.keras import models
from tensorflow.keras import optimizers
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import  to_categorical


#создаем функцию для создания нейросети,апроксимирующей sin,подробнее в примере sin_approximations
def model_sin_approximation(loss ,optimizer):
  model = models.Sequential()
  model.add(layers.Dense(10,activation = 'sigmoid',input_dim=(1)))
  model.add(layers.Dense(1,activation = 'sigmoid')) 
  model.compile(metrics=['accuracy'],loss=loss, optimizer=optimizer)
  return model 

#создаем функцию для сощдания нейросетеи,классифцирующей цифры mnist ,подробнее в примере Mnist
def model_mnist(optimizer):
  model = models.Sequential()
  model.add(layers.Dense(64,activation = 'relu',input_shape=(28*28,))) 
  model.add(layers.Dense(10,activation = 'softmax'))
  model.compile(optimizer=optimizer,loss='categorical_crossentropy',metrics=['accuracy'])
  return model


#Создаем ункцию для создания графиков
def print_res(res):
  for i in res:
    plt.plot(i.history['loss'])
    plt.title('Model loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
  plt.legend(['SGD', 'RMSproop',"adagrad","adadelta","adam"], loc='upper left')
  plt.show()
  

#создаем массив функций оптимизации
optim=[optimizers.SGD(lr=0.01),
       optimizers.RMSprop(learning_rate=0.01, rho=0.9),
       optimizers.Adagrad(learning_rate=0.01),
       optimizers.Adadelta(learning_rate=0.01, rho=0.95),
       optimizers.Adam(learning_rate=0.01, beta_1=0.9, beta_2=0.999, amsgrad=False)]

# создаем датасет
x=np.linspace(0,2,1500) # иницианализируем необходимое количество точек,в указаном отрезке,подробнее тут: https://numpy.org/devdocs/reference/generated/numpy.linspace.html
y = np.sin(x) # получаем массив значений синусов от массива x


#Создаем массив моделей и добавляем модели в массив моделей
models_sin = []
for i in optim:
  models_sin.append(model_sin_approximation("mse",i))
#Создаем массив ответов и обучаем,добавляя результат(резулльтат =  loss) в массив результатов
res = []
for i in models_sin:
  res.append(i.fit(x,y,epochs= 20,batch_size=20,verbose=1))
```

```python
#Рисуем график
print_res(res)
```

```python
#Создаем массив моделей и добавляем модели в массив моделей
models_sin = []
for i in optim:
  models_sin.append(model_mnist(i))
  
  
  
  

  
#готовим датасет цифр 
(train_images, train_labels), (test_images, test_labels) = mnist.load_data() 
train_images = train_images.reshape((60000,28*28))
train_images = train_images.astype('float32')/255
test_images = test_images.reshape((10000,28 * 28))
test_images = test_images.astype('float32')/255
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)





#Создаем массив ответов и обучаем,добавляя результат(резулльтат =  loss) в массив результатов
res = []
for i in models_sin:
  res.append(i.fit(train_images,train_labels,epochs= 20,batch_size=20,verbose=1))
```

```python
print_res(res)
```
